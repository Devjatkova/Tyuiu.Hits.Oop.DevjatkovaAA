Описание задачи:

Создать базовый класс Product в C#, который будет представлять информацию о
продуктах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.


Требования к базовому классу Product:

• Атрибуты: Название (Name), Цена (Price), Производитель (Manufacturer).

• Методы:
o GetInfo(): метод для получения информации о продукте в виде строки.
o Discount(): метод для применения скидки к цене продукта.
o Display(): метод для отображения информации о продукте на экране.


Требования к производным классам:
1. Электроника (Electronics): Должен содержать дополнительные атрибуты,
такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть
переопределен для добавления логики учета гарантийного срока при
применении скидки.
2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как
Размер (Size). Метод Display() должен быть переопределен для добавления
информации о размере при отображении информации о продукте.
3. Книги (Books) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен
быть переопределен для включения информации об авторе в описании
продукта.


#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

```
public interface IDiscount
{
    void Discount();
}

public interface IDisplay
{
    void Display();
}


public class Product : IDiscount, IDisplay //реализация множественного наследования
{
    public string Name { get; set; }
    public int Price { get; set; }
    public string Manufacturer { get; set; }
    public double NewPrice;
 
    public Product(string name, int price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
    }

    public virtual string GetInfo()
    {
        return $"Товар: {Name}, его цена: {Price}, производитель: {Manufacturer}";
    }

    public virtual void Discount()
    {
        NewPrice = Price - (Price * 0.05);
        Console.WriteLine($"На товар {Name} действует скидка в 5%, новая цена товара {NewPrice}");
    }

    public virtual void Display()
    {
        Console.WriteLine($"Информация о товаре {Name} выведена на экран");
    }
}


public class Electronics : Product //реализация простого наследования + добавление нового атрибута и метода
{
    public string WarrantyPeriod { get; set; }
    public int ServiceLife { get; set; } //новый атрибут

    public Electronics(string name, int price, string manufacturer, string warrantyPeriod, int serviceLife) : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
        ServiceLife = serviceLife;
    }

    public override void Discount()
    {
        Console.WriteLine($"Гарантийный срок на товар {Name} составляет {WarrantyPeriod}");
    }

    public void Life() //новый метод
    {
        Console.WriteLine($"Cрок службы товара {Name} - {ServiceLife} лет");
    }
}


public class Clothing: Product //реализация простого наследования + добавление нового атрибута и метода
{
    public int Size { get; set; }
    public string Color { get; set; } //новый атрибут

    public Clothing(string name, int price, string manufacturer, int size, string color) : base(name, price, manufacturer)
    {
        Size = size;
        Color = color;
    }

    public override void Display()
    {
        Console.WriteLine($"Размер товара {Name} - {Size}");
    }

    public void ColorClothing() //новый метод
    {
        Console.WriteLine($"Цвет товара {Name} - {Color}");
    }

    public void Resizing(Clothing clothing)
    {
        clothing.Size -= 2;
    }
}

public class Dress : Clothing //реализация сложного наследования
{
    public string Cut {get; set;}

    public Dress (string name, int price, string manufacturer, int size, string color, string cut) : base(name, price, manufacturer, size, color)
    {
        Cut = cut;
    }

    public void CutDress()
    {
        Console.WriteLine($"Крой платья - {Cut}");
    }
}


public class Books: Product //реализация простого наследования + добавление нового атрибута и метода
{
    public string Author { get; set; }
    public int YearOfPublication { get; set; } //новый атрибут

    public Books(string name, int price, string manufacturer, string author, int yearOfPublication) : base(name, price, manufacturer)
    {
        Author = author;
        YearOfPublication = yearOfPublication;
    }

    public override string GetInfo()
    {
        return $"Товар: {Name}, его цена: {Price}, автор: {Author}";
    }

    public void PriceIncrease(Books book)
    {
        book.Price += 50;
    }

    public void Year() //новый метод
    {
        Console.WriteLine($"Год издания книги - {YearOfPublication}");
    }
}


var Sour_cream = new Product("Молоко тюменское", 65, "ООО «Тюменьмолоко»");
Console.WriteLine(Sour_cream.GetInfo());
Sour_cream.Discount();
Sour_cream.Display();

Console.WriteLine("");

var Telephone = new Electronics("Смартфон Xiaomi Redmi Note 8 2021 64 ГБ синий", 15990, "Xiaomi", "12 месяцев", 5);
Console.WriteLine(Telephone.GetInfo());
Telephone.Discount();
Telephone.Life();

Console.WriteLine("");

var Shirt = new Clothing("Рубашка в клетку", 4500, "ЕМКА", 50, "Красный");
Console.WriteLine(Shirt.GetInfo());
Shirt.Display();
Shirt.ColorClothing();
Shirt.Resizing(Shirt); // Взаимодействие объектов между собой
Console.WriteLine("Изменения информации о товаре");
Shirt.Display();

Console.WriteLine("");

var myDress = new Dress("Платье", 1500, "jchensj", 48, "Розовый", "Блузон");
Console.WriteLine(myDress.GetInfo());
myDress.Display();
myDress.ColorClothing();
myDress.CutDress();

Console.WriteLine("");

var Book = new Books("Обмануть вселенную", 605, "Эксмо", "Макс Максимов", 2021);
Console.WriteLine(Book.GetInfo());
Book.Year();
Book.PriceIncrease(Book); // Взаимодействие объектов между собой
Console.WriteLine("Изменения информации о товаре");
Console.WriteLine(Book.GetInfo());
```
