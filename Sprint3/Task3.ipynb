Описание задачи:

Создать базовый класс Product в C#, который будет представлять информацию о
продуктах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.


Требования к базовому классу Product:

• Атрибуты: Название (Name), Цена (Price), Производитель (Manufacturer).

• Методы:
o GetInfo(): метод для получения информации о продукте в виде строки.
o Discount(): метод для применения скидки к цене продукта.
o Display(): метод для отображения информации о продукте на экране.


Требования к производным классам:
1. Электроника (Electronics): Должен содержать дополнительные атрибуты,
такие как Гарантийный срок (WarrantyPeriod). Метод Discount() должен быть
переопределен для добавления логики учета гарантийного срока при
применении скидки.
2. Одежда (Clothing): Должен содержать дополнительные атрибуты, такие как
Размер (Size). Метод Display() должен быть переопределен для добавления
информации о размере при отображении информации о продукте.
3. Книги (Books) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Автор (Author). Метод GetInfo() должен
быть переопределен для включения информации об авторе в описании
продукта.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

<h2 style="color:DodgerBlue">Реализация:</h2>

```
using System;
using System.Collections.Generic;

public class Product
{
    public string Name { get; set; }
    public int Price { get; set; }
    public string Manufacturer { get; set; }
    public double NewPrice;
 
    public Product(string name, int price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
    }

    public virtual string GetInfo()
    {
        return $"Товар: {Name}, его цена: {Price}, производитель: {Manufacturer}.";
    }

    public virtual void Discount()
    {
        NewPrice = Price - (Price * 0.05);
        Console.WriteLine($"На товар {Name} действует скидка в 5%, новая цена товара {NewPrice}.");
    }

    public virtual void Display()
    {
        Console.WriteLine($"Информация о товаре {Name} выведена на экран.");
    }

    public virtual void Sell() // Новый метод для generic классов
    {
        Console.WriteLine("Товар продаётся в магазине.");
    }
}

public class Electronics : Product // полиморфизм с перекрытием, реализуемый через наследование с переопределением методов
{
    public string WarrantyPeriod { get; set; }

    public Electronics(string name, int price, string manufacturer, string warrantyPeriod) : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
    }

    public override void Discount() // переопределение метода
    {
        Console.WriteLine($"Гарантийный срок на товар {Name} составляет {WarrantyPeriod}.");
    }

    public override void Sell() // переопределение метода
    {
        Console.WriteLine("Электроника продаётся в магазине.");
    }
}

public class Clothing: Product // полиморфизм с перекрытием, реализуемый через наследование с переопределением методов
{
    public int Size { get; set; }

    public Clothing(string name, int price, string manufacturer, int size) : base(name, price, manufacturer)
    {
        Size = size;
    }

    public override void Display() // переопределение метода
    {
        Console.WriteLine($"Размер товара {Name} - {Size}.");
    }

    public override void Sell() // переопределение метода
    {
        Console.WriteLine("Одежда продаётся в магазине.");
    }
}

public class Books: Product // полиморфизм с перекрытием, реализуемый через наследование с переопределением методов
{
    public string Author { get; set; }

    public Books(string name, int price, string manufacturer, string author) : base(name, price, manufacturer)
    {
        Author = author;
    }

    public override string GetInfo() // переопределение метода
    {
        return $"Товар: {Name}, его цена: {Price}, автор: {Author}.";
    }

    public override void Sell() // переопределение метода
    {
        Console.WriteLine("Книга продаётся в магазине.");
    }


    // Реализация перегрузки методов
    // Метод BookCover(), принимающий string
    public void BookCover(string color)
    {
        Console.WriteLine($"Обложка книги {color} цвета.");
    }

    // Метод BookCover(), принимающий bool
    public void BookCover(bool cover)
    {
        if (cover)
        {
            Console.WriteLine("Книга в твёрдом переплёте.");
        }
        else
        {
            Console.WriteLine("Книга в мягком переплёте.");
        }
    }
}


// Реализация generic классов
// Определение генерического класса ProductCollection<T>
public class ProductCollection<T> where T : Product
{
    private List<T> _products = new List<T>();

    public void Add(T product)
    {
        _products.Add(product);
    }

    public void Remove(T product)
    {
        _products.Remove(product);
    }

    public void SellProducts()
    {
        foreach (var product in _products)
        {
            product.Sell(); // Вызов метода Sell() для каждого товара
        }
    }
}

Product Sour_cream = new Product("Молоко тюменское", 65, "ООО «Тюменьмолоко»");
Console.WriteLine(Sour_cream.GetInfo());
Sour_cream.Discount();
Sour_cream.Display();

Console.WriteLine("");

Electronics Telephone = new Electronics("Смартфон Xiaomi Redmi Note 8 2021 64 ГБ синий", 15990, "Xiaomi", "12 месяцев");
Console.WriteLine(Telephone.GetInfo());
Telephone.Discount();

Console.WriteLine("");

Clothing Shirt = new Clothing("Рубашка в клетку", 4500, "ЕМКА", 50);
Console.WriteLine(Shirt.GetInfo());
Shirt.Display();

Console.WriteLine("");

Books Book = new Books("Обмануть вселенную", 605, "Эксмо", "Макс Максимов");
Console.WriteLine(Book.GetInfo());
Book.BookCover("жёлтого");
Book.BookCover(true);

Console.WriteLine("");

ProductCollection<Product> products = new ProductCollection<Product>();

products.Add(Sour_cream);
products.Add(Telephone);
products.Add(Shirt);
products.Add(Book);

products.SellProducts();
```
